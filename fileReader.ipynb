```python
import os
```

```python
def read_jssps(address):
    jssps = []
    files = os.listdir(address)
    for i in range(len(files)):
        file_name = files[i]
        if file_name.endswith(".txt"):
            file = open(address + '/'+file_name)
            jssp = []
            for line in file:
                def ArrtoInt(arr):
                    iArr = []
                    for i in arr:
                        iArr.append(int(i))
                    return iArr
        
                jssp.append(ArrtoInt(line.split(', ')))
            jssps.append({"id":file_name.split('_')[1],"jssp": jssp})
            
    return jssps
```
